```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv')
df
```

```python
# copia del df para para hacer la limpieza

df_clean = df.copy()
df_clean
```

```python
df_clean.drop(1309,axis=0,inplace=True)
```

```python
df_clean
```

```python
#1 
df_clean['age']= round(df_clean.age.fillna(df_clean.age.mean()),1)
df_clean
```

```python
df_clean['age'].isna()
```

```python
#2 Si embarked es na llenar el campo con S
df_clean['embarked'] = df_clean.embarked.fillna('S')
df_clean
```

```python
df_clean['sex']
```

```python
#3 cambiar los alfanumericos de sex a numero 

df_clean['sex']= df_clean['sex'].replace( { 'female': 1, 'male': 0 })
df_clean
```

```python
df_clean['sex']= df_clean['sex'].astype(int)
```

```python
#4 cambiar los alfanumericos de embarked a numero
df_clean['embarked']= df_clean['embarked'].replace({ 'S': 0, 'C': 1, 'Q': 2 })
df_clean
```

```python
#5 Creando la nueva colmna
df_clean['child'] = df_clean['age'].astype(float)
df_clean
```

```python
df_clean['child']
```

```python
#sustituyendo
df_clean.loc[(df_clean.child <16), 'child']=1
df_clean.loc[(df_clean.child >=16), 'child']=0
df_clean['child']

```

```python
#6 Eliminar columnas 
df_clean.drop(['ticket','cabin','name'],axis=1)

```
